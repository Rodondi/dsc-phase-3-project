```python
# Read the data from the "imdb.title.basics.csv.gz" file into a Pandas DataFrame named 'imdb_title_basics'.
imdb_title_basics = pd.read_csv("zippedData/imdb.title.basics.csv.gz")
                                
# Display the DataFrame 'imdb_title_basics', showing the first 5 and last 5 rows to inspect the data.
imdb_title_basics
```

* 'primary_title' - This column also has 146,144 non-null entries; no missing values. The data type is 'object,' (strings); the primary titles of the movies.

```python
# Read the data from the file "imdb.title.ratings.csv.gz" into a Pandas DataFrame named 'imdb_title_ratings.'
imdb_title_ratings = pd.read_csv("zippedData/imdb.title.ratings.csv.gz")

# Display the DataFrame 'imdb_title_ratings', showing the first 5 and last 5 rows to inspect the data
imdb_title_ratings
```

```python
# Print the summary of the DataFrame 'imdb_title_ratings', including information about non-null counts and data types.
imdb_title_ratings.info()
```

* The DataFrame 'imdb_title_ratings' contains a total of 73,856 rows and 3 columns.
* 'tconst' - This column has 73,856 non-null entries; no missing values; The data type is 'object,' (strings).'tconst' likely represents the IMDb title identifier for each movie.
* 'averagerating' - This column also has 73,856 non-null entries; no missing values; The data type is 'float64,' (float); 'averagerating' represents the average ratings given to each movie.
* 'numvotes' - This column has 73,856 non-null entries; no missing values; The data type is 'int64,' (integer); 'numvotes' represents the number of votes received by each movie.

## b) Cleaning and Merging the Datasets

```python
# Remove the release year information (in brackets) from the 'title' column and strip any leading/trailing spaces.
bom['title'] = bom['title'].str.replace(r'(\(\d{4}( re-release| re-issue| Restoration)?\))', '').str.strip()

# Remove commas from the 'foreign_gross' column to convert it to numeric format.
bom["foreign_gross"] = bom["foreign_gross"].str.replace(',', '')

# Fill missing values in 'domestic_gross' and 'foreign_gross' with 0.
bom['domestic_gross'] = bom['domestic_gross'].fillna(0)
bom['foreign_gross'] = bom['foreign_gross'].fillna(0)

# Convert 'foreign_gross' and 'domestic_gross' columns to numeric data types.
bom['foreign_gross'] = bom['foreign_gross'].astype(float)
bom['domestic_gross'] = bom['domestic_gross'].astype(int)

# Calculate the 'total_gross' by adding 'domestic_gross' and 'foreign_gross'.
bom['total_gross'] = bom['domestic_gross'] + bom['foreign_gross']

# Display the updated 'bom' DataFrame after data preparation and cleaning.
bom
```

```python
# Print the summary of the updated 'bom' DataFrame, including information about non-null counts and data types
bom.info()
```

The output confirms that the cleaning steps were successful in updating the 'bom' DataFrame. The missing values were appropriately handled, and the data types were converted to suitable numeric types where necessary, making the data ready for further analysis.

```python
# Merge imdb_title_basics and imdb_title_ratings on 'tconst'
imdb_merged = pd.merge(imdb_title_basics, imdb_title_ratings, on='tconst', how='inner')

# Merge bom dataframe with the imdb_merged dataframe on 'title'
merged_data = pd.merge(bom, imdb_merged, left_on='title', right_on='primary_title', how='inner')

# Drop duplicate columns and unnecessary columns
merged_data.drop(columns=['original_title', 'primary_title'], inplace=True)

# Display the merged dataframe
merged_data
```

```python
# Print the summary of the merged DataFrame 'merged_data', including information about non-null counts and data types.
merged_data.info()
```

* The DataFrame contains 3485 rows (entries) and 12 columns
* The 'studio', 'runtime_minutes' and 'genres' columns contain some missing values, as indicated by the difference between the number of non-null values and the total number of rows (3485 entries).

```python
# Mode imputation for 'studio' and 'genres'
merged_data['studio'].fillna(merged_data['studio'].mode()[0], inplace=True)
merged_data['genres'].fillna(merged_data['genres'].mode()[0], inplace=True)

# Mean imputation for 'runtime_minutes'
mean_runtime = merged_data['runtime_minutes'].mean()
merged_data['runtime_minutes'].fillna(mean_runtime, inplace=True)

# Verify if there are any remaining missing values
print(merged_data.isnull().sum())
```

```python
# Display the merged dataframe
merged_data
```

## Q1: What is the trend of Total Gross Revenue over the years?

```python
# Group the data by year and calculate the mean total gross revenue for each year
revenue_by_year = merged_data.groupby('year')['total_gross'].mean().reset_index()

# Plot the trend of total gross revenue over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='year', y='total_gross', data=revenue_by_year, marker='o')
plt.title("Trend of Total Gross Revenue Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Gross Revenue (in billions)")
plt.grid(True)
plt.show()
```

Based on the resulting line plot, the trend of total gross revenue over the years shows an overall positive pattern. 

The graph illustrates that the total gross revenue has generally increased over time, indicating a growth in the movie industry's revenue generation. 

It can be observed that there may have been fluctuations in revenue in certain years, but the overall trend is upward. 

This suggests that the movie industry has experienced growth and profitability over the years, with higher total gross revenue generated in more recent years compared to earlier years.

This trend of increasing total gross revenue is essential information for Microsoft's new movie studio. 

It indicates a potential positive market scenario, and as a result, Microsoft's studio can be optimistic about the potential success and financial viability of producing movies in the current market.

However, further analysis can be conducted to identify specific factors or movie attributes that contribute to the revenue growth. 

This additional analysis can help in making strategic decisions about the types of movies to produce and the target audience to focus on, further improving the chances of success in the highly competitive movie industry.

## Q2: Which movie studios have the highest average ratings for their films?

```python
# Group the data by 'studio' and calculate the mean average rating for each studio
average_ratings_by_studio = merged_data.groupby('studio')['averagerating'].mean().reset_index()

# Sort the studios in descending order based on their average ratings
average_ratings_by_studio = average_ratings_by_studio.sort_values(by='averagerating', ascending=False)

# Select the top 10 movie studios with the highest average ratings
top_10_studios = average_ratings_by_studio.head(10)

# Plot the top 10 movie studios with the highest average ratings
plt.figure(figsize=(10, 6))
sns.barplot(x='averagerating', y='studio', data=top_10_studios)
plt.title("Top 10 Movie Studios with Highest Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Studio")
plt.grid(True)
plt.show()
```

Based on the resulting bar plot:

We can observe that Trafalgar, NAV, GrtIndia, SHO, BSC, PDA,App., Good Deed, RME and MUBI movie studios received the highest average ratings for their films. This indicates their strong performance in delivering movies that are well-received by the audience.

This information is valuable for Microsoft's new movie studio, as it highlights the studios that have a track record of producing high-rated films. Collaborating or partnering with these top-rated studios could potentially improve the chances of creating successful and well-regarded movies.

However, it's essential to consider other factors such as the genre, target audience, and market trends before making any business decisions. 

Additionally, further analysis can be done to explore the relationship between studio reputation and other aspects that contribute to movie success, leading to more informed and strategic choices for the new movie studio.

## Q3: What are the most common genres of movies that have received high ratings?

```python
# Step 1: Filter movies with high ratings (e.g., average rating above 7.5)
high_rated_movies = merged_data[merged_data['averagerating'] > 7.5]

# Step 2: Extract and count occurrences of each genre from high-rated movies
genres_list = high_rated_movies['genres'].str.split(',')
genres_count = {}
for genres in genres_list:
    for genre in genres:
        genre = genre.strip()
        if genre in genres_count:
            genres_count[genre] += 1
        else:
            genres_count[genre] = 1

# Step 3: Create a visualization to display the top genres with the highest occurrences
top_genres_df = pd.DataFrame({'Genre': list(genres_count.keys()), 'Count': list(genres_count.values())})
top_genres_df = top_genres_df.sort_values(by='Count', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Genre', data=top_genres_df)
plt.title("Most Common Genres of Movies with High Ratings")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.grid(True)
plt.show()
```

Based on the resulting bar plot we can identify the most common genres of movies that have received high ratings (i.e., movies with an average rating above 7.5). 

From the graph, we can observe that Drama, Documentary, Comedy, Action, Adventure, Biography, Romance, Crime, Thriller and Animation genres have the movies with high ratings indicating their popularity and frequency in movies that have been well-received by the audience.

Knowing the most common genres of high-rated movies can be valuable for Microsoft's new movie studio. 

It provides insights into the types of movies that tend to receive positive reviews and audience appreciation. By focusing on these popular genres, the studio can increase the likelihood of producing movies that resonate with viewers and potentially achieve high ratings and success.

However, it's important to note that other factors, such as the plot, cast, and production quality, also play significant roles in determining a movie's success. 

Therefore, a comprehensive analysis of various attributes and their impact on movie ratings can further enhance the studio's decision-making process.

## Q4: Is there a correlation between the runtime of a movie and its average rating?

```python
# Step 1: Extract the necessary data from the merged dataframe
runtime_rating_data = merged_data[['runtime_minutes', 'averagerating']]

# Step 2: Calculate the correlation coefficient between 'runtime_minutes' and 'averagerating'
correlation_coefficient = runtime_rating_data.corr().iloc[0, 1]

# Step 3: Create a scatter plot to visualize the relationship between the two variables
plt.figure(figsize=(8, 6))
sns.scatterplot(x='runtime_minutes', y='averagerating', data=runtime_rating_data)
plt.title("Correlation between Runtime of a Movie and its Average Rating")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()

print("Correlation Coefficient:", correlation_coefficient)
```

Based on the resulting scatter plot and the correlation coefficient of 0.1373, we can observe that there is a weak positive correlation between the runtime of a movie and its average rating.

Each point in the plot represents a movie, with its runtime on the x-axis and its average rating on the y-axis.

A positive correlation between the two variables means that, on average, as the runtime of a movie increases, its average rating also tends to increase. 

However, the correlation coefficient of 0.1373 indicates that the relationship is weak, meaning that the increase in average rating with longer runtime is not very strong.

While there is a slight tendency for longer movies to receive slightly higher average ratings, the effect is not significant enough to suggest a strong relationship between movie runtime and rating. 

Other factors, such as the movie's plot, acting, direction, and genre, are likely to play a more crucial role in determining the movie's rating.

As such, the Micrososft's new Movie Studio should not solely rely on runtime to impact the movie's average rating but should focus on creating compelling and engaging content that resonates with the audience to achieve higher ratings and success.

## Q5: How does the number of votes received by a movie impact its average rating?

```python
# Step 1: Extract the necessary data from the merged dataframe
votes_rating_data = merged_data[['numvotes', 'averagerating']]

# Step 2: Create a scatter plot to visualize the relationship between 'numvotes' and 'averagerating'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='numvotes', y='averagerating', data=votes_rating_data)
plt.title("Impact of Number of Votes on Average Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Average Rating")
plt.grid(True)
plt.show()

# Step 3: Calculate the correlation coefficient between 'numvotes' and 'averagerating'
correlation_coefficient = votes_rating_data.corr().iloc[0, 1]
print("Correlation Coefficient:", correlation_coefficient)
```

Based on the resulting scatter plot and the correlation coefficient of 0.2619, we can observe that there is a positive correlation between the number of votes received by a movie and its average rating.

Each point in the plot represents a movie, with the number of votes it received on the x-axis and its average rating on the y-axis.

A positive correlation between the two variables means that, on average, as the number of votes received by a movie increases, its average rating tends to increase as well. 

Movies that receive more votes are more likely to have higher average ratings.

This positive correlation suggests that popular movies that attract a larger audience and receive more votes tend to be well-received by viewers, leading to higher average ratings. 

The correlation coefficient of 0.2619 indicates that the relationship between the number of votes and average rating is moderate, but it is still strong enough to suggest a meaningful connection.

For Microsoft's movie studio, this finding implies that movies with higher viewer engagement, represented by a larger number of votes, are more likely to achieve higher average ratings. 

Therefore, creating engaging and appealing content that resonates with the audience and encourages participation can have a positive impact on a movie's overall rating and success.
